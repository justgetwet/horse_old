```python
"""
Race on the sqlite3
"""
from datetime import datetime
import numpy as np
import pandas as pd
import sqlite3
from contextlib import closing # 
sqlite3.register_adapter(np.int64, lambda val: int(val))
print("library loading ", datetime.now().strftime("%Y/%m/%d %H:%M:%S"))
```

```python
""" race テーブルの作成 """
with closing(sqlite3.connect("racing.sqlite3")) as conn:
    cur = conn.cursor()
    # cur.execute("Drop Table if exists Race;") # 削除
    Create_Table = """ Create Table if not exists Race (\
    RaceCode Text not Null Primary Key,\
    Date Text, Course Text, R Text, RaceName Text, TrkDist Text,\
    Cond Text, Holding Text, Detail Text, Prize Text); """
    try:
        cur.execute(Create_Table)
    except sqlite3.Error as e:
        print('sqlite3.Error occurred:', e.args[0])
    conn.commit()
    conn.close()
print("code loading ", datetime.now().strftime("%Y/%m/%d %H:%M:%S"))
```

```python
""" Entries 出馬表 """
with closing(sqlite3.connect("racing.sqlite3")) as conn:
    cur = conn.cursor()
    # cur.execute("Drop Table if exists Entries;") # 削除
    Create_Table = """ Create Table if not exists Entries (\
    id Integer not Null Primary Key Autoincrement,\
    RaceCode Text, Bk Integer, Hs Integer, Horse Text,\
    SexAge Text, Weight Real, Jockey Text, Trainer Text,\
    HsWeight Text,Odds Real, Fav Integer, HorseCode Text); """
    try:
        cur.execute(Create_Table)
    except sqlite3.Error as e:
        print('sqlite3.Error occurred:', e.args[0])
    conn.commit()
    conn.close()
print("code loading ", datetime.now().strftime("%Y/%m/%d %H:%M:%S"))
```

```python
""" リーディング """
with closing(sqlite3.connect("racing.sqlite3")) as conn:
    for tbl in ['Jockey', 'Trainer', 'Sire', 'Breeder']:
        cur = conn.cursor()
        # cur.execute(f"Drop Table if exists {tbl};")
        Create_Table = f"Create Table if not exists {tbl} (" + \
        "id Integer not Null Primary Key Autoincrement," + \
        "Rank Integer, Name Text, Rate Real);"
        try:
            cur.execute(Create_Table)
        except sqlite3.Error as e:
            print('sqlite3.Error occurred:', e.args[0])
        conn.commit()
    conn.close()
print("code loading ", datetime.now().strftime("%Y/%m/%d %H:%M:%S"))
```
